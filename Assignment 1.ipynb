# Q20

```python
pd.set_option ('display.max_rows',None)
cars = pd.read_csv('Cars.csv')
cars
```

```python
cars['MPG'].mean()
```

```python
cars['MPG'].median()
```

so the data can be considered normally distributed

```python
from scipy import stats
from scipy.stats import norm
```

```python
# P(MPG > 38)
stats.norm.sf(38,cars.MPG.mean(),cars.MPG.std())
```

```python
# P(MPG < 40)
stats.norm.cdf(40,cars.MPG.mean(),cars.MPG.std())
```

```python
# P(20 < MPG < 50)
stats.norm.cdf(50,cars.MPG.mean(),cars.MPG.std())-stats.norm.cdf(20,cars.MPG.mean(),cars.MPG.std())
```

# Q9

```python
Q9_a = pd.read_csv('Q9_a.csv')
Q9_a
```

```python
Q9_a[['speed','dist']].skew()
```

Since the skewness of speed is between -0.5 & 0.5 the data is nearly symmetrical.

Since the skewness of dist is between between 0.5 & 1 the data is slightly positively skewed.

```python
Q9_a[['speed','dist']].kurt()
```

Negative kurtosis indicates flatter peak and thinner tails. Positive kurtosis indicates sharper peak and thicker tails.

```python
Q9_a.plot.scatter(x='speed',y='dist')
plt.show()
```

From the scatter plot both speed and dist are positively correalted.

```python
Q9_b=pd.read_csv('Q9_b.csv')
Q9_b
```

```python
Q9_b.plot.scatter(x='SP',y='WT')
plt.show()
```

```python
Q9_b[['SP','WT']].skew()
```

Since the skewness of SP is greater than 1 the data must be extremely positively skewed. 

Since the skewness of WT is less than 0.5 the data must be slightly negatively skewed.

```python
Q9_b[['SP','WT']].kurt()
```

kurtosis of 3(SP=2.97) indicates standard normal distribution.

kurtosis less than 3 indicates the distribution is less peaked and has lighter tails.
